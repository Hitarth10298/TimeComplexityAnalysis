```python
%matplotlib inline     
import matplotlib.pyplot as plt            
import numpy as np      
```

BELOW IS THE SAMPLE GRAPH FOR FUNCTION f(n) = 7n^2+6n+5. IT CAN BE CLEARLY SEEN AS THE VALUE OF "n" INCREASES, THE TIME COMPLEXITY INCREASES AS WELL. THE VALUE OF "n" IS SET WITH THE HELP OF LINSPACE. HERE THE VALUE IS 100

```python
n = np.linspace(1, 100)
plt.plot(n, 7 * n * n + 6 * n + 5)    # f(n) = 7n^2+6n+5 for the values of n from 0 to 100. 
plt.show()
```

NOW, LET'S CHECK EITHER OF **LINEAR FUNCTION AND QUADRATIC FUNCTION**, WHICH ONE PERFORMS BETTER.

```python
n = np.linspace(1, 100)    # Here the value of n is set as 100
plt.plot(n, 7 * n * n + 6 * n + 5, label="7n^2+6n+5")         # Quadratic Function = 7n^2+6n+5
plt.plot(n, 7 * n, label="7n")      # Linear Function = 7n
plt.legend(loc='upper left')        # legend will allow us to set the bar which displays what is what
plt.show()
```

NOW, LET'S CHECK HOW THE GRAPH VARIATES FOR THE QUADRATIC FUNCTION WHEN THE VALUE OF "n" IS LARGE AND SMALL.

```python
plot1 = plt.figure(1)
n = np.linspace(0,50)       # Visualizing Quadratic function for larger value of "n"
plt.plot(n, 7 * n**2+6*n+5, label="7n^2+6n+5")
plt.plot(n, 7 * n**2, label="7n^2")
plt.legend(loc='upper left')
plt.title('For large values of n')
plt.show()

plot2=plt.figure(2)
n = np.linspace(0,5)        # Visualizing Quadratic function for smaller value of "n"
plt.plot(n, 7 * n**2+6*n+5, label="7n^2+6n+5")
plt.plot(n, 7 * n**2, label="7n^2")
plt.legend(loc='upper left')
plt.title('For smaller values of n')
plt.show()
```

ANOTHER WAY FOR VISUALIZING THE RESULTS FOR QUADRATIC FUNCTION IS BY DIVIDING BOTH THE FUNCTIONS.

```python
n = np.linspace(1, 100)
plt.plot(n, (7*n**2+6*n+5)/(7*n**2))      # Another way of visualizing it is by plotting ((7n^2 + 6n + 5) / 7n^2). Here, we can observe the graph bending towards 0 as "n" increases.
plt.show()
```

MULTIPLICATIVE CONSTANTS CAN BE IGNORED

```python
n = np.linspace(1, 500)
plt.plot(n, (7*n**2+6*n+5)/(n**2))      # Trying to establish the fact in the same way.
plt.show()
```

PERFORMING ANALYSIS ON POLYNOMIAL FUNCTION. HERE, WE WILL OBSERVE THAT THE FUNCTION WITH HIGHEST DEGREE WILL GROW FASTER.

```python
# Performing the Polynomial function by taking the function as n, n^2 and n^3.
# Here, we will observe that the function with highest degree will grow faster.

plot1 = plt.figure(1)
n = np.linspace(1, 3)     # Taking "n" = 3
plt.plot(n, n, label="n")
plt.plot(n, n * n, label="n^2")
plt.plot(n, n * n * n, label="n^3")
plt.legend(loc='upper left')
plt.title('For smaller n')
plt.show()

plot2 = plt.figure(2)
n = np.linspace(1, 50)      # Taking "n" = 50
plt.plot(n, n, label="n")
plt.plot(n, n * n, label="n^2")
plt.plot(n, n * n * n, label="n^3")
plt.legend(loc='upper left')
plt.title('For larger n')
plt.show()

plot3 = plt.figure(3)
n = np.linspace(1, 1000)      # Taking "n" = 1000
plt.plot(n, n, label="n")
plt.plot(n, n * n, label="n^2")
plt.plot(n, n * n * n, label="n^3")
plt.legend(loc='upper left')
plt.title('For very large n')
plt.show()
```

COMPARING THE RESULTS OF POLYNOMIAL FUNCTION WITH EXPONENTIAL. AS THE VALUE OF "n" INCREASES THE GROWTH OF POLYNOMIAL REDUCES AS COMPARED TO EXPONENTIAL AND VICE-VERSA.

```python
# Any Polynomial grows smaller than the Exponential, only when the value of "n" increases. If the value of "n" is smaller Polynomial grows faster.


plot1 = plt.figure(1)
n = np.linspace(1, 5)  # For smaller "n". We will observe that the polynomial will grow faster.
plt.plot(n, n ** 4, label="n^4")
plt.plot(n, 2 ** n, label="2^n")
plt.legend(loc='upper left')
plt.title('For smaller n')
plt.show()

plot2 = plt.figure(2)
n = np.linspace(1, 20)    # As the value of "n" is increased, the exponential will outperform polynomial.
plt.plot(n, n ** 4, label="n^4")
plt.plot(n, 2 ** n, label="2^n")
plt.legend(loc='upper left')
plt.title('For relatively larger n')
plt.show()

plot3 = plt.figure(3)
n = np.linspace(1, 1000)    # As the value of "n" is increased drastically, the growth of exponential will be very high.
plt.plot(n, n ** 100, label="n^100")
plt.plot(n, 2 ** n, label="2^n")
plt.legend(loc='upper left')
plt.title('For larger n')
plt.show()
```

COMPARING RESULTS BETWEEN POLYLOGARITHM AND POLYNOMIAL. WE WILL BE ABLE TO SEE THAT POLYNOMIAL WILL OUTPERFORM POLYLOGARITHM. 

```python
# Most of the time, any Polylogarithm grows slower than the Polynomial

n = np.linspace(1, 20)    # Taking "n" as 20
plt.plot(n, n, label="n")   # Polynomial = n
plt.plot(n, np.log(n), label="log n")     # Polylogarithm = log n
plt.legend(loc='upper left')
plt.show()
```

LET'S TAKE THE VALUE OF "n"=50

```python
# When the value of "n" = 50, it can be observed that Polylogarithm outperforms Polynomial.

n = np.linspace(1, 50) 
plt.plot(n, n ** .5, label="n^0.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
plt.title('Small n')
plt.show()
```

```python
# Taking "n" = 10^7, Polylogarithmic grows faster than Polynomial.

n = np.linspace(1, 10**7)
plt.plot(n, n ** .5, label="n^0.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
plt.title('Large n')
plt.show()
```

```python
# But, "n" = 10^8 will decrease the growth of Polylogarithmic.

n = np.linspace(1, 10**8)
plt.plot(n, n ** .5, label="n^0.5")
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.legend(loc='upper left')
plt.title('Even Larger n')
plt.show()
```

COMPARING RESULTS BETWEEN POLYLOGARITHMIC AND QUADRATIC

```python
# Polylogarithmic grows slower than the Quadratic for both smaller and larger values of "n".

plot1 = plt.figure(1)
n = np.linspace(0,50)
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.plot(n, 7 * n**2, label="7n^2")
plt.legend(loc='upper left')
plt.title('For large values of n')
plt.show()

plot2=plt.figure(2)
n = np.linspace(0,5)
plt.plot(n, np.log(n) ** 3, label="(log n)^3")
plt.plot(n, 7 * n**2, label="7n^2")
plt.legend(loc='upper left')
plt.title('For smaller values of n')
plt.show()
```
